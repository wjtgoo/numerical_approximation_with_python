```python
import numpy as np
import matplotlib.pyplot as plt
def lagrange(x,X,Y):
    if len(X) != len(Y):
        raise ValueError("输入的插值节点X变量与Y变量长度不对应！")
    if type(x)==int:
        x = [x]
    Y = np.array(Y)
    n = len(X)
    y = []
    # 求所有待估计点
    for xi in x:
        # 定义l保存l_1,l_2,...l_n
        l = np.ones(n)
        for i in range(n):
            for ii in range(n):
                if i != ii:
                    l[i] = l[i]*(xi-X[ii])/(X[i]-X[ii])
        yi = Y@l
        y.append(yi)
    y = np.array(y)
    return y
    
```

```python
x = np.arange(4,10,step=0.5)
y = lagrange(x,[1,4,9,16],[1,2,3,4])
```

```python
print(y)
```

```python
plt.plot(x,y,label='Lagrange')
plt.plot(x,np.sqrt(x),label='Origin')
#plt.scatter(x[7],y[7],s=30.,c='r')
plt.legend(loc='best')
plt.show()
```

## y = x^2 插值

```python
X = np.linspace(-5,5,10)
Y = X**2
```

```python
x = np.linspace(-20,20,100)
y = lagrange(x,X,Y)
```

```python
num_cha = 50
range_cha = 1
X = np.linspace(-range_cha,range_cha,num_cha)
Y = X**2
x = np.linspace(-20,20,100)
y = lagrange(x,X,Y)
y_real = x**2
error_max = abs(y-y_real).max()
for num in range(50):
    range_cha=20
    for rg in range(20):
        X = np.linspace(-(range_cha-rg),range_cha-rg,num_cha-num)
        Y = X**2
        y = lagrange(x,X,Y)
        error_max = abs(y-y_real).max()
        if error_max>0.1:
            print('插值节点数：',num_cha-num,'a：',range_cha-rg)
            break
```
